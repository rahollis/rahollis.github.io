2. What is a bifurcation?

3. What can a bifurcation diagram tell you about a mathematical model?

#### Answers:

```python
def fibonacci(N):
    sequence = [0,1]
    for i in range(N-1):
        sequence.append(sequence[-2] + sequence[-1])
    return sequence
```

```python
fibonacci(10)
```
