```python
pattern = "DG4 DB24 DG2 K24 DG24 R4"
colors = {"DG" : [0,100,0],"DB" : [28,0,112],"K" : [16,16,16],"R" : [200,0,0]}
```

```python
import numpy as np
```

```python
n = 300
```

```python
dg4 = np.zeros((n,4,3))
db24 = np.zeros((n,24,3))
dg2 = np.zeros((n,2,3))
k24 = np.zeros((n,24,3))
dg24 = np.zeros((n,24,3))
r4 = np.zeros((n,4,3))
```

```python
dg4[:,:,1] = 100
db24[:,:,0] = 28
db24[:,:,2] = 112
dg2[:,:,1] = 100
k24[:,:,:] = 16
dg24[:,:,1] = 100
r4[:,:,0] = 200
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(db24/255)
plt.show()
```

```python
vert = np.concatenate([dg4,db24,dg2,k24,dg24,r4],axis = 1)
vert = np.concatenate([vert]*(n//vert.shape[1]+1),axis = 1)[:,:n,:]/255
```

```python
plt.imshow(vert)
plt.show()
```

```python
horiz = np.transpose(vert,axes=(1,0,2))
plt.imshow(horiz)
plt.show()
```

```python
basket = np.zeros((100,100,3))
for i in range(100):
    for j in range(100):
        if (i+j)%2 == 0:
            basket[i,j,:] = horiz[i,j,:]
        else:
            basket[i,j,:] = vert[i,j,:]
plt.imshow(basket)
plt.show()
```

```python
tartan = np.zeros((n,n,3))
for i in range(n):
    for j in range(n):
        if (i-j+1)%4 < 2:
            tartan[i,j,:] = horiz[i,j,:]
        else:
            tartan[i,j,:] = vert[i,j,:]
plt.imshow(tartan)
plt.show()
```
